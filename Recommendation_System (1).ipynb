```python
# it converts strings that "look" like Python lists or dictionaries
# into actual Python lists or dictionaries.

import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
# Instead of taking every single person who worked on the movie,
#  you are extracting only the top 3 actors.

def convert_cast(text):
    l = []
    counter = 0

    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
import ast

def convert_crew(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
ps = PorterStemmer()

def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return ' '.join(l)
```

```python
import numpy as np
import pandas as pd
import pickle
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. CRITICAL: Reset index and clean tags
# This ensures movie 'Spider-Man' is at the same row index in the matrix
new_df = new_df.reset_index(drop=True)
new_df['tags'] = new_df['tags'].str.lower().fillna('')

# 2. Vectorization (Optimized for Large Datasets)
cv = CountVectorizer(max_features=5000, stop_words='english')
# We keep it as a sparse matrix to save RAM
vector = cv.fit_transform(new_df['tags'])

# 3. Memory-Safe Similarity Calculation (Top 50 Indices)
def create_similarity_indices(vector_matrix, top_k=50, chunk_size=500):
    n_rows = vector_matrix.shape[0]
    # We store only the INDICES of the top 50 movies to save 99% RAM
    similarity_indices = np.zeros((n_rows, top_k), dtype=np.int32)

    print(f"Starting calculation for {n_rows} movies...")
    for i in range(0, n_rows, chunk_size):
        end = min(i + chunk_size, n_rows)
        # Calculate similarity for a chunk against all movies
        sim_chunk = cosine_similarity(vector_matrix[i:end], vector_matrix)

        # Get indices of top_k most similar items
        partitioned = np.argpartition(sim_chunk, -top_k, axis=1)[:, -top_k:]

        # Sort results within the chunk so the best match is first
        for j in range(end - i):
            row_idx = i + j
            top_indices = partitioned[j]
            scores = sim_chunk[j, top_indices]
            similarity_indices[row_idx] = top_indices[np.argsort(scores)[::-1]]

        print(f"Processed {end}/{n_rows} movies...")
        del sim_chunk # Clear RAM

    return similarity_indices

# Execute calculation
similarity = create_similarity_indices(vector)

# 4. Multi-Functional Explore/Recommend Function
def explore_movies(query_type, value=None):
    """
    query_type: 'recommend', 'language', 'rating', or 'genre'
    value: Movie title, language code (ml, en), or genre name
    """
    # A. RECOMMENDATION (Based on Similarity)
    if query_type == 'recommend':
        # Robust search: strip spaces and ignore case
        search_term = str(value).strip().lower()
        match = new_df[new_df['title'].str.lower().str.strip() == search_term]

        if not match.empty:
            idx = match.index[0]
            print(f"\nMovies similar to '{new_df.iloc[idx].title}':")
            # Get pre-calculated top indices (skip first as it's the movie itself)
            movie_indices = similarity[idx][1:6]
            for i in movie_indices:
                print("-", new_df.iloc[i].title)
        else:
            print(f"Movie '{value}' not found. Try checking the exact title.")

    # B. FILTER BY LANGUAGE (e.g., 'ml' for Malayalam)
    elif query_type == 'language':
        results = new_df[new_df['original_language'] == value].head(10)
        print(f"\nTop {value.upper()} movies:")
        for title in results['title']:
            print("-", title)

    # C. HIGH RATING
    elif query_type == 'rating':
        # Only consider movies with a significant number of votes
        results = new_df[new_df['vote_count'] > 100].sort_values('vote_average', ascending=False).head(10)
        print("\nHighly Rated Movies (with >100 votes):")
        for title in results['title']:
            print("-", title)

    # D. FILTER BY GENRE
    elif query_type == 'genre':
        # Robust genre search
        results = new_df[new_df['genres'].astype(str).str.contains(value, case=False)].head(10)
        print(f"\nTop {value} movies:")
        for title in results['title']:
            print("-", title)

# 5. Save Artifacts (Fixed 'artifacts' directory name)
if not os.path.exists('artifacts'):
    os.makedirs('artifacts')

pickle.dump(new_df, open('artifacts/movie_list.pkl', 'wb'))
pickle.dump(similarity, open('artifacts/similarity.pkl', 'wb'))

print("\nProcessing Complete! Testing output...")
explore_movies('recommend', 'Spider-Man 3') # Test with a title likely in both datasets
```
